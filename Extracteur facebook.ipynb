```python
def lien_comment(soup):
    url_comments=[]


    i=0
    #on recupere tout les article de la recherche 
    for article in soup.find_all('article'):             
        #Récuperation des commentaires de chaque articles
        span=article.find_all('span')


        for sp in span:
            if not sp.get("id"):
                id_span='none'

            else:
                id_span=sp.get("id")
                id_=article.find('span',{"id":sp.get("id")})

            #on recupere l'url en lien avec les commentaires


                testcontent=[]
                for test in id_:
                    #print(test.get('href'))
                    if isinstance(test, NavigableString):
                        testcontent.append('none')
                    else:
                        testcontent.append(test.get('href'))

                    url_test=testcontent[0]

        url_comments.append(url_test)


        #on recupere les commentaire de chaque articles 
    nbr_article=len(url_comments)
    #print(nbr_article)
    return(url_comments)
```

```python
#il faut recuperer le test1 (saveid)
def recuperer_texte_propri_comment(soup,saveid,url_comments):
    commentaire_article=[]
    for compteur in range(len(url_comments)):
       #id du des comdu poste

        id_poste=saveid[compteur]
        try:
            #ouvre l'url des commentaire du poste
            url_like="https://mbasic.facebook.com"+url_comments[compteur]
            driver.get(url_like)
            html = driver.page_source
            commentairehtml=BeautifulSoup(html)

            #initiation d'un dictionnaire
            list_commentaire={}


            #on recherche les div des commetaires qui commencent par les trois premier chiffre de l'id
            comment_div_id=[]
            for i, test in enumerate(commentairehtml.find_all('div')):
                div_id=test.get("id")
                #print(div_id)
                if not div_id:
                    next
                else:
                    if div_id[:3]==id_poste[:3]:
                        comment_div_id.append(div_id)


            #on recupere la class des balise
            if not comment_div_id:
                continue
            else:
                balise=commentairehtml.find('div',{"id":comment_div_id[0]})
                class_balise=balise["class"]

            for i,test in enumerate(commentairehtml.find_all('div',{"class":class_balise})):

                    if not test.find('h3'):
                           next
                    else:
                        prorietaire_comment=test.find('h3').text

                        commentaire=test.find('h3').findNext('div').text
                    list_commentaire["proprietaire "+str(i)]=prorietaire_comment
                    list_commentaire["commentaire "+str(i)]=commentaire
            #print(list_commentaire)
            commentaire_article.append(list_commentaire)

        except WebDriverException:
            print("page down")
    return commentaire_article
```

```python
def recuperer_contenu_post(soup):
    content_search=[]

    i=0
    #on recupere tout les article de la recherche 
    for article in soup.find_all('article'):
        content=[]
        
        #print(article)
        #buffer=article
       # on recupere tout les texte de la recherche 
        p=article.find_all('p')
        img=article.find_all('img')
        span=article.find_all('span')



        for text in p:
            content.append(text.text)
        content_search.append(content)
    
    content_searchfinal=[]
    for element in content_search:
        if not element:
            content_searchfinal.append(["pas de commentaire"])
        else:
             content_searchfinal.append(element)
    return content_searchfinal
```

```python
def recuperer_lien_image(soup):
    content_search=[]
    img_search=[]
    
    #on recupere tout les article de la recherche 
    for article in soup.find_all('article'):
        
        content_img=[]

    
        img=article.find_all('img')


        for source in img:
            content_img.append(source['src'])

        img_search.append(content_img)
    return img_search
```

```python
def recuperer_nbr_like(soup):
    saveid=[]
    like_content=[]
    #on recupere tout les article de la recherche 
    for article in soup.find_all('article'):
       
        sp_content=[]
        span=article.find_all('span')


        #Recupération des likes
        for sp in span:
        #print(sp.get("id"))
            if not sp.get("id"):
                id_span='none'
            else:
                id_span=sp.get("id")
                #print(id_span[:5])
                if id_span[:5]=='like_':
                    sp_content.append(sp.get("id"))
                    #print(sp_content)
                    
                    number_like=article.find('span',{"id":sp.get("id")})
                     #On en profite pour recupere l'id de l'article
                    #print(number_like['id'].split('_'))
                    idbuffer=number_like['id'].split('_')
                    saveid.append(idbuffer[1])

                    #id_article.append(id_)
                    
                    
                    
                    
                    number_like=number_like.find('a').text
                   
                    number_like=number_like.replace(u'\xa0','')

                    like_content.append(number_like)
    #print(saveid)
    return(like_content,saveid)


```

```python
def dataframe_result(post_content,image_link,numberlike,commentaires):
            
    #On mets toute les data dans des df
    #df content
    post_content=pd.DataFrame(data=post_content)
    post_content['contents'] = post_content[post_content.columns[0:]].apply( 
        lambda x: ','.join(x.dropna().astype(str)),
        axis=1)
    post_content=pd.DataFrame(data=post_content['contents'])




    #df url image
    image_link=pd.DataFrame(data=image_link)
    image_link=pd.DataFrame(data=image_link[0])
    image_link.columns=["url_img"]


    #df number like
    numberlike=pd.DataFrame(numberlike)
    numberlike.columns=["number_of_like"]



    #df commentaire
    commentaires=pd.DataFrame(commentaires)

    image_link.reset_index(drop=True, inplace=True)
    numberlike.reset_index(drop=True, inplace=True)
    commentaires.reset_index(drop=True, inplace=True)

    dataset_scrapp_facebook=pd.concat([post_content,image_link,numberlike,commentaires],axis=1)
    
    return dataset_scrapp_facebook


```

```python
post_content=recuperer_contenu_post(soup)
image_link=recuperer_lien_image(soup)
numberlike,id_=recuperer_nbr_like(soup)
url_comments=lien_comment(soup)
commentaires=recuperer_texte_propri_comment(soup,id_,url_comments)
dataframe_scrap=dataframe_result(post_content,image_link,numberlike,commentaires)

```
